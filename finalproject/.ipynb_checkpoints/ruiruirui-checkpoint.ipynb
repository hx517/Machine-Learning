```python
import pandas as pd
import numpy as np
import pylab as plt
import time
from datetime import datetime
% pylab inline
```

```python
df=pd.read_csv('ML_data.csv')
```

```python
df.head(2)
```

## Divide time into one hour for clustering
## these divided time will be classified in the previous hour, add date

```python
df['datetime'] = df["CMPLNT_FR_DT"].map(str) + " " + df["CMPLNT_FR_TM"]    
```

```python
df.head(2)
```

```python
df = df.sort_values('datetime')
df
```

```python
df = df.drop(df.index[:20])
df
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['datetime'] = pd.to_datetime(df.datetime.dt.date) + pd.to_timedelta(df.datetime.dt.hour, unit='H')
df.head()
```

## Divide time into one hour for clustering
## these divided time will be classified in the previous hour, add date

```python
#weather.assign(rounded_dt=weather.datetime.dt.round('H'))
df['time']=pd.to_timedelta(df.datetime.dt.hour, unit='H')
df.index = range(len(df))
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(  
    df['datetime'],   
    format='%Y/%m/%d'  
) 
```

### filter for only year 2017

```python
df = df[(df['date'] >='20170101') ]
df.head(1)
```

```python
df.index = range(len(df))
df
```

```python
df['weekday'] = df['date'].dt.dayofweek
df
```
